```python
import json
import pandas
import requests
```

```python
# 觀察檔案
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
data = json.loads(r.text)
for d in data:
    titles = d['title']
    likes = d['likeCount']
    print('%s的【按讚人數】為%3d' % (titles,likes))
```

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
datas = len(pandas.read_json(r.text))
print('此API一次會回傳%s筆資料' % datas)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
# 熱門文章
cols_name = ['標題', '貼文時間', '留言人數', '按讚人數']
content = pandas.read_json(r.text)
contents = content[['title','createdAt','commentCount','likeCount']]
contents.columns = cols_name
contents
```
